### Git Clone

##### Kopier clone-linken fra github.com

##### Paste inn og trykk clone


### Git Branch
"Stå" i mappen i browseren.



##### Velg branch
Om du velger fra en som starter med "origin/" vil den forsøke å kopiere den lokalt fra github.com



### Git Add
Legger til filer som skal være med i "saven" - "committen" \


### Git Commit

Utfører "selve saven", med en tittel og en kommentar. \


### Git Push

Sender endringene på "branchen" til "origin", fra det du la i "committen" \


##### Brukernavn er brukernavnet ditt på github.com, og passordet er "personal access token", se oppskrift her:
https://wiki.ssb.no/display/MAS2/Bruk+av+github+i+statistikkproduksjon#Brukavgithubistatistikkproduksjon-Jupyter-integrasjonmotGithub \


### Git Pull

Hent ned endringer på branches fra origin. \
Feiler ofte fordi du har lokale endringer som ikke er lagret, eller som må "merges"... \
