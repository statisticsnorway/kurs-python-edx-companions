### Git Clone

##### Kopier clone-linken fra github.com

##### Paste inn og trykk clone


### Git Branch
"St" i mappen i browseren.



##### Velg branch
Om du velger fra en som starter med "origin/" vil den forske  kopiere den lokalt fra github.com



### Git Add
Legger til filer som skal vre med i "saven" - "committen" \


### Git Commit

Utfrer "selve saven", med en tittel og en kommentar. \


### Git Push

Sender endringene p "branchen" til "origin", fra det du la i "committen" \


##### Brukernavn er brukernavnet ditt p github.com, og passordet er "personal access token", se oppskrift her:
https://wiki.ssb.no/display/MAS2/Bruk+av+github+i+statistikkproduksjon#Brukavgithubistatistikkproduksjon-Jupyter-integrasjonmotGithub \


### Git Pull

Hent ned endringer p branches fra origin. \
Feiler ofte fordi du har lokale endringer som ikke er lagret, eller som m "merges"... \
