### Command og Edit-kontekst

```python
# Viktig å kommentere koden sin, til fordel for nestemann
if 1: print('x')
# Det siste som presenteres som ett "objekt" i en celle vil implisitt printe...
# Dette er en "list-comprehension", som produserer en liste med en for-loop.
[i for i in range(5)]

# Men ikke om man gjør noe annet etterpå (uncomment linja under)...
#if 2: print('y')
```

### Mest brukte hotkeys / shortcuts

- Esc - for å ut av en celle til "command mode"
- Enter (numpad) - gå inn i editmode
- ctrl + s - save

- ctrl + enter   - Kjør cellen du står i
- shift + enter  - Kjør cellen du står i og gå til neste

- m - konverter cellen til markdown
- y - konverter cellen til kode

I commandmode:
- c - copy
- v - paste
- dd - delete
- a - legg inn celle foran
- b - legg inn celle bak
- piltaster - bla opp og ned i commandmode

- tab - indent/autocomplete
- shift tab - remove indent
