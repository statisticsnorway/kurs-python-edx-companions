##### Seperasjon av problematikk knyttet til at "feature-utvikling" kan kræsje kode
Forskjellige features, eller forskjellige folk, kan ha hver sin "branch". \
Selv om din branch kræsjer fordi du har gjort noe galt med koden du skriver, så vil den ikke kræsje for andre. \
Om ikke du har pushet feilaktig kode, og de igjen har pullet fra dette.

##### Seperasjon av ansvar, og tydelig arbeidsfordeling
Du jobber her, jeg der. Men vi kan også bytte, men da må vi snakke sammen. \
Denne branchen er for å arbeide med dette. \
If you break it, you fix it.

##### Muligheter for å godkjenne og ha "status" på kodebiter
- Koden kan ligge lenge i "utvikling" før den legges i en branch som regnes som "klar" (Dev -> Prod)
- Før dette kan man ha lange diskusjoner om hvordan ting bør fungere osv.

##### Mindre krav om at alle må ha full oversikt over hele prosjektet
Denne branchen er for å jobbe med kun dette, og du trenger bare å vite om "grenseområdene".

## Hvordan gjøres dette?

- Lokale kopier av koden (ikke "origin")
- Kontroll over hva som skal "være med" (add)
- Logg og beskrivelse av endringer som er gjort (commit)
- Når man skal sende tilbake kilden (push)
- Når man skal hente fra kilden (pull)
- Hvordan overlappende kode skal kombineres (merge)
- Mulighet for å påtvinge godkjenning av prosesser (feks. review før merge av en pull request)
